```python
import pandas as pd
from pathlib import Path
```

```python
files = Path("../data/").glob("*.csv")

all_dfs = []
for file in files:
    all_dfs.append(pd.read_csv(file))

df = pd.concat(all_dfs)
print(len(df))
```

```python
df = df[df['battle_type'] == 2]
print(len(df))
```

```python
p1_wins = df[df['winner'] == 1].groupby('p1_polaris_id').size()
p2_wins = df[df['winner'] == 2].groupby('p2_polaris_id').size()
```

```python
total_wins = p1_wins.add(p2_wins, fill_value=0)
print(total_wins)
```

```python
p1_matches = df['p1_polaris_id'].value_counts()
p2_matches = df['p2_polaris_id'].value_counts()
total_matches = p1_matches.add(p2_matches, fill_value=0)
print(total_matches)
```

```python
overall_wr = total_wins / total_matches
overall_wr = overall_wr.fillna(0)
print(overall_wr)
```

```python
p1_char_wins = df[df['winner'] == 1].groupby(['p1_polaris_id', 'p1_chara_id']).size()
print(p1_char_wins)
```

```python
p2_char_wins = df[df['winner'] == 2].groupby(['p2_polaris_id', 'p2_chara_id']).size()
print(p2_char_wins)
```

```python
p1_char_wins.index.names = ['polaris_id', 'chara_id']
p2_char_wins.index.names = ['polaris_id', 'chara_id']
total_char_wins = p1_char_wins.add(p2_char_wins, fill_value=0)
print(total_char_wins)
```

```python
p1_char_matches = df.groupby(['p1_polaris_id', 'p1_chara_id']).size()
print(p1_char_matches)
```

```python
p2_char_matches = df.groupby(['p2_polaris_id', 'p2_chara_id']).size()
print(p2_char_matches)
```

```python
p1_char_matches.index.names = ['polaris_id', 'chara_id']
p2_char_matches.index.names = ['polaris_id', 'chara_id']
total_char_matches = p1_char_matches.add(p2_char_matches, fill_value=0)
print(total_char_matches)
```

```python
char_wr = total_char_wins / total_char_matches
char_wr = char_wr.fillna(0)
print(char_wr)
```

```python
matchup_wins = df[df['winner'] == 1].groupby(['p1_chara_id', 'p2_chara_id']).size()
print(matchup_wins)
```

```python
matchup_total = df.groupby(['p1_chara_id', 'p2_chara_id']).size()
print(matchup_total)
```

```python
matchup_wr = matchup_wins / matchup_total
print(matchup_wr)
```

```python
overall_wr = (total_wins / total_matches).fillna(0)
overall_wr = overall_wr.reset_index()
overall_wr.columns = ['polaris_id', 'overall_wr']
print(overall_wr.head())
```

```python
char_wr = (total_char_wins / total_char_matches).fillna(0)
char_wr = char_wr.reset_index()
char_wr.columns = ['polaris_id', 'chara_id', 'char_wr'] 
print(char_wr.head())
```

```python
matchup_wr = (matchup_wins / matchup_total).fillna(0)
matchup_wr = matchup_wr.reset_index()
matchup_wr.columns = ['p1_chara_id', 'p2_chara_id', 'matchup_wr']
print(matchup_wr.head())
```
